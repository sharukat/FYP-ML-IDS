```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('Datasets/csv_result-KDDTrain+.csv')
df.drop('id',axis=1,inplace=True)
```

```python
def categorical_data_plot():
    df_grp1 = pd.DataFrame(df.groupby(['class','protocol_type'])['count'].count())
    df_grp2 = pd.DataFrame(df.groupby(['class','flag'])['count'].count())
    
    a_protcol = []
    n_protcol = []
    a_flag = []
    n_flag = []
    for i in range(0,3):
        a_protcol.append(df_grp1['count'].values[i])
        n_protcol.append(df_grp1['count'].values[i+3])
        
    for i in range(0,11):
        a_flag.append(df_grp2['count'].values[i])
        if not i==10:
            n_flag.append(df_grp2['count'].values[i+11])
    n_flag.append(0)
    
    fig, (ax1,ax2) = plt.subplots(1,2,figsize=(20,10))
    width =0.4
    
    pl1=ax1.bar(np.arange(len(a_protcol)), a_protcol, width=width,label='anomaly')
    pl2=ax1.bar(np.arange(len(n_protcol))+ width, n_protcol, width=width , label='normal')
    ax1.set_xticks(np.arange(len(a_protcol)) + width / 2)
    ax1.set_xticklabels( labels=['icmp', 'tcp', 'udp'],fontsize=14,rotation=45)
    ax1.legend(loc='best')
    ax1.set_yticks([])
    ax1.set_title('protocol_type count in each class',fontsize=18)
    def autolabel1(rects,ax):
        for rect in rects:
            height = rect.get_height()
            ax.text(rect.get_x() + rect.get_width()/2., 1.01*height,
                    '%d' % int(height),ha='center', va='bottom',fontsize=14)
    autolabel1(pl1,ax1)
    autolabel1(pl2,ax1)
    
    
    pl3=ax2.bar(np.arange(len(a_flag)), a_flag, width=width,label='anomaly')
    pl4=ax2.bar(np.arange(len(a_flag))+ width, n_flag, width=width , label='normal')
    ax2.set_xticks(np.arange(len(a_flag)) + width / 2)
    ax2.set_xticklabels(labels=['OTH','REJ','RSTO','RSTOS0','RSTR','S0','S1','S2','S3','SF','SH'],fontsize=14,rotation=45)
    ax2.legend(loc='best')
    ax2.set_yticks([])
    ax2.set_title('flag count in each class',fontsize=18)
    def autolabel2(rects,ax):
        for rect in rects:
            height = rect.get_height()
            ax.text(rect.get_x() + rect.get_width()/2., 1.02*height,
                    '%d' % int(height),ha='center', va='bottom',rotation=90,fontsize=14)
    autolabel2(pl3,ax2)
    autolabel2(pl4,ax2)
    
    
```

```python
categorical_data_plot()
```
